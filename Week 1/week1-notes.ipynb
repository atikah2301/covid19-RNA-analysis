### Floating point numbers

Floating point numbers (floats, for short) are real numbers written in decimal form. Python stores floats and integers in different ways, in particular, floats are stored with fixed precision of about 16 digits.

```python
#square root of two
2.0**0.5
```

Python understands scientific notation

```python
1.0e-4
```

```python
1.23e4
```

The computer representation of floating point numbers does not use the decimal number system, but the binary one. When the binary number gets converted to decimal (and vice versa), rounding of the result to 16 digits can result in small errors in the output, as in the following computation, which should have produced an exact result of $534.302208$.

```python
332*1.609344
```

### Complex Numbers

Python can do calculations with complex numbers. A complex number is expressed with using `j` for $j=\sqrt{-1}$ as the imaginary unit. (Yes, I know you are used to $i^2=-1$, but $j$ is sometimes also used, for example in engineering mathematics textbooks.) Alternatively, you can use `complex()` with two arguments $x$ and $y$ to give you $x+i\,y$.

```python
complex(3,2)
```

```python
3+2j
```

```python
(1+2j)*(1+1j)
```

Note that we needed to write `1+1j`, and that `1+j` will produce an error (this has to do with the use of variables, covered in the next section). Note also that the output of a complex calculation is itself of type complex, even if the result looks real

```python
(1+2j)*(1-2j)
```

```python
type(5+0j)
```

## Variables

### The assignment operator

The assignment operator `=` assigns a value to a variable. For example, to assign the integer $3$ to the variable `x` we write `x=3`:

```python
x=3
```

Note that this assignment does not produce any output. However, once this assignment has been done, `x` will evaluate to the value $3$.

```python
x
```

We can use the variable `x` to evaluate expressions involving `x`.

```python
1+x+x**2
```

```python
x/(1+x)
```

### Variable names to use

You can use any sequence of letters, digits, and underscores (`_`) to create a variable name, as long as it does not start with a digit. To keep code readable, it makes sense to have names that give some indication of the intended use. In the example above, I chose `x` for a variable used to compute further expressions, just as in $y=f(x)$ the independent variable is denoted by $x$. 

To keep code readable in this module, we will use [naming conventions](http://www.python.org/dev/peps/pep-0008/#naming-conventions) `lower_case_with_single_underscores` variable names and single lowercase letter variable names.

This naming convention makes code as the following easy to read:

```python
#compute the distance from Edinburgh to London in kilometres
distance_miles=332
km_per_mile=1.609344
distance_km=distance_miles*km_per_mile
distance_km
```

## Showcasing some advanced examples

The following examples are supposed to get you interested. They show how easy it is to use Python to help you with mathematical problems

```python
#first, import the needed modules (details will come later)
import numpy
import matplotlib.pyplot as plt # note the abbreviation to plt
from mpl_toolkits import mplot3d # note the import of part of a module only
```

We can graph one or several functions

```python
xvals = numpy.arange(0, 2*math.pi, 0.01) # define xvals
yvals = numpy.sin(xvals) # generate yvals by evaluating the function sin at x values
plt.plot(xvals, yvals) # create a line plot with yvals against xvals
plt.show() # display the resulting figure
```

```python
xvals = numpy.arange(0, 2*math.pi, 0.01)
yvals = numpy.sin(xvals)
plt.plot(xvals, yvals)
yvals2 = numpy.cos(xvals) # add new yvals by evaluating the function cos at x values
plt.plot(xvals, yvals2) # add second line plot with new yvals
plt.show()
```

We can generate randomly distributed data and produce a histogram

```python
# generating data of a normal distribution and plotting the histogram
plt.hist(numpy.random.randn(1000000), bins=50);
plt.xlabel('x')
plt.ylabel('count')
plt.show()
```

And finally, a more advanced example. Here we want to understand the function 
$$f(x,y)=(1-\frac x2+x^5+y^3) \exp(-x^2-y^2)$$
We will do a three-dimensional plot and a contour plot

```python
# define the function - I haven't told you yet how 'def' works, but don't worry
def f(x,y):
    return(1-x/2+x**5+y**3)*numpy.exp(-x**2-y**2)
```

```python
# create an (x,y) grid
n = 256
x = numpy.linspace(-3,3,n)
y = numpy.linspace(-3,3,n)
X,Y = numpy.meshgrid(x,y)
Z = f(X,Y)
```

Having prepared an (x,y) grid and evaluated the function at all points in the grid, we can now produce the desired plots

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
plt.contourf(X, Y, f(X,Y))
plt.show()
```
