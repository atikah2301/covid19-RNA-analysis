**[2.1] [5 marks]**

- Define a function, $mapping$(x)
- input is a list of characters
- output is a list of integers
- $A \mapsto -2; \ C \mapsto -1; \ G \mapsto 1; \ T \mapsto 2\;.$
- e.g. $x_L=[A,T,T,A,C,G]$ is mapped into $x_N=[-2,2,2,-2,-1,1]$

---


````python
def mapping(x): 
    for letter,number in [('A',-2),('C',-1),('G',1),('T',2)]: 
        x[:] = [char if char != letter else number for char in x]
    return x

# x represents a list of characters, such as str2 from [1.2]
# loop over each mapping pair of a letter to a number, which I have given as a list of tuples
# for each mapping, update the list x, using list comprehension that loops over every character in x
# it replaces each instances of the specified letter with its corresponding number mapping
# the list comprehension edits the list x by using successive conditional statements

# using the first iteration of the for loop as an example:
# if any element (char) in x is not the specific letter, in this case A, then it is left unchanged. So, C,G,T remain as such.
# if an element (char) in x IS the specific letter, in this case A, then it is replaced with the number -2
# once all instances of A have been replaced, the next iteration of the for loop repeats the process, but for the next mapping
# once the for loop is done, every element of x would have been mapped to a number, so return the fully updated list x

mapped_virus_data = mapping(str2.copy())
print(mapped_virus_data[:10]) 

# testing my code by printing a slice of str2 after it has been mapped by mapping(x)
# only using a copy() of str2 to avoid editing the data, as i'll be using slices of str2 later for more testing purposes
````

**[2.2] [5 marks]** Implement a function called $walker$(x) that, given a list $x$ of N numbers [x(0), x(1), x(2),...,x(N-1)], outputs a "walk list" $y=[y(0), y(1), ..., y(N)]$, defined recursively as: 
$$y(0) = 0,\\
y(n) = y(n-1) + x(n-1), \ \text{if} \ 0<n\leq N$$

- define $walker$(x)
- input is a list of integers, x, of size N
- output is a list of recursively defined integers, y, of size N+1
- y is essentially a list of cumulative sums

---


````python
def walker(x): 
    y = [0]
    for i in range(len(x)):
        #print(y[i],x[i])
        y.append(y[i]+x[i])
        #print(y)
    return y

# x is a list of number integer values, such as those given by the mapping(x) function from [2.1]
# start the code by initialising out list of outputs, y, with y(0)=0
# iterate over the indices of the two lists, x and y, using i values in the range of (0,N) since len(x)=N
# append new values to the output list, y, using the recursive formula provided
# once the loops ends, return the fully constructed list of y values

walker_test = [1,-23,2,-22,3]
walker(walker_test) 

# test on a list of length 5, output is a list of sums of length 6, as expected
````

````python
# Before implementing linear_fit(z), I have extracted some of the steps for formatting the input z into its own function
# I named the intermediate function unzip(), and used it within the code for linear_fit(z)
 
def mean(x):
    """returns the arithmetic mean value as a float, for a list of numbers x"""
    # although a mean(x) function is already build into python, it did not seem to work each time, so I created my own
    return sum(x)/len(x)
    
def unzip(z):
    """this function takes a list of paired numbers, z, and splits it into a pair of lists, like the inverse of the zip function
        for example, z = [(10,2),(6,15),(3,7)] becomes ([10,6,3],[2,15,7])"""
    xvalues = [] # empty lists, each for storing x and y values later
    yvalues = []
    for x,y in z: # let the first element of each pair in z be x, let the second element be y
        xvalues.append(x)
        yvalues.append(y) # add each element of the pair to separate lists
    return xvalues, yvalues

def linear_fit(z):
    """the function takes a list of paired numbers, z, 
        computes the coefficients of their least-squares fit,
        and returns them as a pair of floats"""    
    x,y = unzip(z)
    a = (mean([x[i]*y[i] for i in range(len(x))]) - mean(x)*mean(y)) / \
        (mean([x[i]*x[i] for i in range(len(x))]) - mean(x)*mean(x))
    # I used list comprehension to create the <x_i y_i> and <x_i^2> terms in the formula for LSEs given
    b = mean(y) - a*mean(x)
    return a,b

import numpy as np
def linear_fit_test(z):
    a,b=np.polyfit(*zip(*z), 1)
    return a,b

# my code was written using only in-built python functions
# compare my results to that given by an existing function in numpy

test_z = [(10,2),(6,15),(3,7)]
print(linear_fit_test(test_z))
print(linear_fit(test_z))

# outputs are largely the same to a great number of significant figures, as I hoped
# because this indicates that the computations are almost identical
````

````python
def linear_trend(z):
    """This function takes any list of ordered y values (observations) from a data set,
        numbers each element with its corresponding x value to create an ordered pair, 
        creates a line of best fit using least-squares fit on nthis set of co-ordinates,
        and returns a list of the fitted (i.e expected y) values"""
    z = list(enumerate(z)) # let the list of numbers z=[z0,z1,..] become z=[(0,z0),(1,zp)..], i.e. taking the form (p,zp)
    a,b = linear_fit(z) # use linear_fit() from [2.3] to get the least-squares fit coefficients, a,b
    fitted_vals = [a*p + b for p in range(len(z))] # create a list for each p=0,1,2.. with the fitted values
    return fitted_vals
    
test_trend = [34,15,8,10,7]
linear_trend(test_trend) # test the function on a list of numbers
````

Use the function *freq()* to plot a frequency histogram (bargraph plot) of the virus sequence, where the frequency of each nucleotide should appear as a blue bar, and the x axis should depict the four nucleotides A,C,G and T from left to right in alphabetical order.

Clarification on QMplus: [Q 3.1] You cannot use any module or function which is pre-defined in a module in the definition of freq(). Then, when you need to use freq() to make some plots, of course you can use pre-defined functions to actually plot.

````python
def freq1(dataset):
    """The function takes a list and creates a dictionary, where the the keys are the unique elements of the list
        and the values are the corresonding frequency.
        Input: list of data e.g. ['A','C','G','T',...]
        Output: dictionary of data and its frequencies e.g. {'A': 8908, 'C': 5484, 'G': 5853, 'T': 9575}"""
    counter = {}
    for data in sorted(dataset):
        if data not in counter:
            counter[data] = 0
        counter[data] += 1
    return counter

# Use sorted() on the input list to ensure the nucleotides appear in alphabetical order left to right on the plot later
# Loop through each element of the list, checking if that unique element is a key in the dictionary called counter
# If it isn't a key yet, set up a key with value 0
# For each list element we loop over, regardless of whether it already has a key or not,..
# ...add 1 to the value of the key the element corresponds to, to count it

one_gram = freq1(str2.copy()) 
print(one_gram)

# get copy of the RNA sequence as a list of charcters
# named the frequency dictionary one_gram to align with the 2-grams in the next question

plt.bar(one_gram.keys(),one_gram.values(), color="#E0BBE4")
plt.xlabel("Nucelotide")
plt.ylabel("Frequency")
plt.title("Frequency of each nucleotide in Virus RNA sample")

plt.show()
````

**[3.2] [8 marks]** 
A so-called 2-gram is defined as 2 consecutive letters. 
It is easy to see that in a sequence of  ð‘  letters, we can count at most ð‘âˆ’1 unique 2-grams
For sequences composed by letters whose alphabet is of size 4 (ACGT) there are a total of  2^4=16  possible 2-grams: AA,AC,AG,AT,...,TT.
Compute and plot a bar graph plot of the frequency of the 16 different 2-grams in the virus sequence

---

````python
def freq2(dataset):
    """the function and creates a dictionary, 
        where the the keys are unique pairs of consequtive elements of the input list
        and the values are the corresonding frequency of that pair.
        Input: list of data e.g. ['A','C','G','T',...]
        Output: dictionary of 2-grams and their frequencies e.g. {'AT': 2303, 'TT': 3207, 'TA': 2371,...}"""
    d = {}
    for i in range(len(dataset)-1): # iterate over the full length of the dataset
        gram2 = dataset[i] + dataset[i+1] # construct the 2-gram
        if gram2 not in d:
            d[gram2] = 0 # if the 2-gram is not already in the dictionary as a key, create a key for it
        d[gram2] += 1 # count frequencies by adding 1 to key's value for each occurance of the 2-gram
    return d

two_gram = freq2(str2.copy()) 
print(two_gram)
print(f"\nThe number of 2-grams accounted for is {len(two_gram)}, as expected.")

# get copy of the RNA sequence as a list of chaarcters
# two_gram is the frequency dictionary of all the 2-grams
# check that the freq2() function does actually account for all 16 2-grams

plt.bar(two_gram.keys(),two_gram.values(), color="#D291BC")
plt.xlabel("2-gram of RNA")
plt.ylabel("Frequency")
plt.title("Frequency of each 2-gram in Virus RNA Sample")
plt.show()
````

**[3.3] [10 marks]** Let $N(ij)$ be the frequency (that is, the number of occurrences) of the 2-gram 'ij' in the virus sequence, for $i,j=A,C,G,T$.
The transition matrix ${\bf T}=\{T_{ij}\}$ of a given sequence is defined such that the general term $T_{ij} = N(ij)/N_{tot}(i),$
where $N_{tot}(i)$ is the frequency of letter 'i' showing up in the sequence. By construction, all rows of $\bf T$ should sum up one.

(A) Compute the $4\times 4$ transition matrix $\bf T$ of the virus sequence. Print the Matrix

(B) Display the matrix as a heatmap using *seaborn*.

(C) Confirm that the largest eigenvalue of $\bf T$ is one, and give the associated eigenvector $v$. Check your results by computing ${\bf T}v$.

---

````python
## PART A ##

# T is a matrix of the frequencies of each 2-gram relative to the frequency of each A,C,G,T value
# Each element of T corresponds to the 2-grams in their alphabetrical order
# So, T[0,0] gives the frequency for AA, T[0,1] is for AC,... T[3,3] is for TT

# Start with an empty 4x4 Matrix as a 2D Array and with an empty list, which will later carry the elements of matrix T
# all2grams is a list of all of the 2-grams sorted in alphabestical order

T = np.zeros((4,4))
T_values = []
all2grams = sorted(list(two_gram.keys()))

# loop through each 2-gram, calculate the relative frequency using the formula, and append these to T_values to store

for k in all2grams:
    T_values.append(two_gram[f'{k}'] / one_gram[f'{k[0]}'])
    
# Reverse the storage list T_values,
# so that when we use pop(), the elements are being assigned to entries in T in the correct order
# pop() is retrieveing then deleting the last value in the list T_values, saving us the trouble of iterating through the list

#T_values = [i for i in range(1,17)] # test that the code works by packing numbers in ascending order into the matrix
T_values.reverse()    
for x, y in np.ndindex(T.shape):
    T[x,y] = T_values.pop()
    
print(T)                

# Visual representation of how the transition matrix is structured

# A [[0. 0. 0. 0.]
# C  [0. 0. 0. 0.]
# G  [0. 0. 0. 0.]
# T  [0. 0. 0. 0.]]
#     A  C  G  T
````

````python
## PART B ##

nucleo = list(one_gram.keys())
seaborn.heatmap(T, xticklabels = nucleo, yticklabels = nucleo, annot = True, cmap="YlGnBu")
# darker colour on the heatmap means higher relative frequency
# the heat map grid corresponds to each entry in the 4x4 transition matrix
````

**[3.4] [8 marks]** Define a function called $deviation(x)$. The input of the function is a walk list $x$. This function performs the following computation:
* It iteratively considers all possible sublists $x_k=[x(0), ..., x(k)]$, where $k=1,2,4,\ldots$ runs through powers of $2$.
* For each possible value of $k$, the function computes $\Delta(k) = max(x_k) - min(x_k)$. 
* The function $deviation$ finally returns a list of the form $[(k,\Delta(k))]$ for all possible values of $k$. 
* That final list is called the *scaling* (this is important for next questions below)

For the case $x=[0,1,2,3,4,5,6,7,8]$ you should get $[(1, 1), (2, 2), (4, 4), (8, 8)]$, whereas for the case $x=[1,1,1,1]$ you should get $[(1,0),(2,0)]$.

---

````python
def deviation(x):
    k = 1
    scaling = []
    while k < len(x):
        x_k = x[:k+1]
        delta = max(x_k) - min(x_k)
        scaling.append((k,delta))
        k *= 2
    return scaling

# k_limit helps define how long of a list of powers of 2 we will need
# Ensures highest power in the k list is not greater than the length of the list x used as input to the deviation() function
# Loop through each k value in ascending order, and create sublists called x_k of length k
# delta is the largest difference between values in the current sublist
# scaling is a list of these delta values being paired with the value of k they occured at

print([1 for i in range(4)])
print(deviation([1 for i in range(4)]))
print()
print([i for i in range(9)])
print(deviation([i for i in range(9)]))

# testing my code on the example given in the question
````
